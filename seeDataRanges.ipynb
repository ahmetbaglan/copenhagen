```python
def createGoodFileInhabitants():
    
    newD = pd.DataFrame({'depthInfo' : [], 'val':[], 'year' : [],  'district' :[]})
    for i in range(len(data)):
        a =  data.iloc[i]['no_couples_w_kids']
        b = data.iloc[i]["no_couples_wo_kids"]
        c = data.iloc[i]["no_single_w_kids"]
        d = data.iloc[i]["no_single_wo_kids"]
        
        year = data.iloc[i]['aar']
        district = data.iloc[i]['district'].replace('-',' ')
#         print district
        newD = newD.append({'depthInfo':district+'-couple-With Kids' ,'val':a, 'year' : year,  'district' : district}, ignore_index = True)
        newD = newD.append({'depthInfo':district+'-couple-Without Kids' ,'val':b, 'year' : year,  'district' : district},ignore_index= True)
        newD = newD.append({'depthInfo':district+'-single-With Kids' ,'val':c, 'year' : year,  'district' : district},ignore_index= True)
        newD = newD.append({'depthInfo':district+'-single-Without Kids' ,'val':d, 'year' : year,  'district' : district},ignore_index= True)
        
    return newD
```

```python
def createGoodFileAge():
    
    newD = pd.DataFrame({'depthInfo' : [], 'val':[], 'year' : [],  'district' :[]})
    for i in range(len(data)):
        
        a =  data.iloc[i]['no_age_0_5']
        b = data.iloc[i]["no_age_6_17"]
        c = data.iloc[i]["no_age_18_29"]
        d = data.iloc[i]["no_age_30_39"]
        e = data.iloc[i]["no_age_40_49"]
        f = data.iloc[i]["no_age_50_64"]
        g = data.iloc[i]["no_age_over_65"]
        
        year = data.iloc[i]['aar']
        district = data.iloc[i]['district'].replace('-',' ')

        newD = newD.append({'depthInfo':district+'-0~5' ,'val':a, 'year' : year,  'district' : district}, ignore_index = True)
        newD = newD.append({'depthInfo':district+'-6~17' ,'val':b, 'year' : year,  'district' : district},ignore_index= True)
        newD = newD.append({'depthInfo':district+'-18~29' ,'val':c, 'year' : year,  'district' : district},ignore_index= True)
        newD = newD.append({'depthInfo':district+'-30~39' ,'val':d, 'year' : year,  'district' : district},ignore_index= True)
        newD = newD.append({'depthInfo':district+'-40~49' ,'val':e, 'year' : year,  'district' : district},ignore_index= True)
        newD = newD.append({'depthInfo':district+'-50~64' ,'val':f, 'year' : year,  'district' : district},ignore_index= True)
        newD = newD.append({'depthInfo':district+'-over 65' ,'val':g, 'year' : year,  'district' : district},ignore_index= True)
        
    return newD
```

```python
def createGoodFileApartment():
    newD = pd.DataFrame({'depthInfo' : [], 'val':[], 'year' : [],  'district' :[]})
    for i in range(len(data)):
        a =  data.iloc[i]['no_apts_before_1900']
        b = data.iloc[i]["no_apts_1900_1929"]
        c = data.iloc[i]["no_apts_1930_1949"]
        d = data.iloc[i]["no_apts_1950_1979"]
        e = data.iloc[i]["no_apts_after_1980"]

        
        year = data.iloc[i]['aar']
        district = data.iloc[i]['district'].replace('-',' ')

        newD = newD.append({'depthInfo':district+'-before 1900' ,'val':a, 'year' : year,  'district' : district}, ignore_index = True)
        newD = newD.append({'depthInfo':district+'-1900~1929' ,'val':b, 'year' : year,  'district' : district},ignore_index= True)
        newD = newD.append({'depthInfo':district+'-1930~1949' ,'val':c, 'year' : year,  'district' : district},ignore_index= True)
        newD = newD.append({'depthInfo':district+'-1950~1979' ,'val':d, 'year' : year,  'district' : district},ignore_index= True)
        newD = newD.append({'depthInfo':district+'-after 19080' ,'val':e, 'year' : year,  'district' : district},ignore_index= True)
        
    return newD
    
```

```python
def createGoodFileIncome():
    newD = pd.DataFrame({'depthInfo' : [], 'val':[], 'year' : [],  'district' :[]})
    for i in range(len(data)):
        a =  data.iloc[i]['no_low_income']
        b = data.iloc[i]["no_middle_income"]
        c = data.iloc[i]["no_high_income"]

        
        year = data.iloc[i]['aar']
        district = data.iloc[i]['district'].replace('-',' ')

        newD = newD.append({'depthInfo':district+'-Low Income' ,'val':a, 'year' : year,  'district' : district}, ignore_index = True)
        newD = newD.append({'depthInfo':district+'-Middle Income' ,'val':b, 'year' : year,  'district' : district},ignore_index= True)
        newD = newD.append({'depthInfo':district+'-High Income' ,'val':c, 'year' : year,  'district' : district},ignore_index= True)
        
    return newD
    
```
